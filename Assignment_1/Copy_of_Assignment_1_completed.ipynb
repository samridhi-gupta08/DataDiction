Question-Write a Python function named slope_of_cubic that calculates the slope of a cubic polynomial based on user input. The function should:
1. Accepts two arguments:

* A tuple containing 4 real numbers which are the coefficients of the cubic term, quadratic term, linear term, and constant term respectively.

* A real number which is the value of x at which the slope has to be calculated.

2. Returns the slope of the cubic polynomial at the given value of x.

```python
def slope_of_cubic(coefficients, x):
    a, b, c, d = coefficients
    return 3 * a * x ** 2 + 2 * b * x + c
```

#### Generating the dataset

Run the following load command to get the 2-Dim iris dataset. Do the following operations on it and please dont edit this cell

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
```

```python
# Write a Pandas function to drop rows that contain a missing value.

import pandas as pd
df_iris_2d = pd.DataFrame(iris_2d)
df_iris_2d_no_missing = df_iris_2d.dropna()
print(df_iris_2d_no_missing.shape)
```

```python
# Write a Pandas function to find the mean, median and standard deviation of column 3.

col3 = df_iris_2d_no_missing.iloc[:, 2]
mean_col3 = col3.mean()
median_col3 = col3.median()
std_col3 = col3.std()
print(mean_col3)
print(median_col3)
print(std_col3)
```

```python
# Write a Pandas function to find the correlation between the first two columns.

first_col = df_iris_2d_no_missing.iloc[:, 0]
second_col = df_iris_2d_no_missing.iloc[:, 1]
correlation = first_col.corr(second_col)
print(correlation)
```

```python
# Write a Pandas function to create a new column for volume in iris_2d, where volume is (pi x petallength x sepal_length^2)/3
# Hints:
# 1. Introduce new dimension to match iris_2d's
# 2. Add the new column

sepal_length = df_iris_2d_no_missing.iloc[:, 0]
petal_length = df_iris_2d_no_missing.iloc[:, 2]
volume = (np.pi * petal_length * (sepal_length ** 2)) / 3
df_iris_2d_no_missing['volume'] = volume
print(df_iris_2d_no_missing.head())
```

# Q4 Write a Pandas function to bin the petal length (3rd) column of iris_2d to form a text array, such that if petal length is:
- Less than 3 –> ‘small’
- 3-5 –> ‘medium’
- greater than equal to 5 –> ‘large’

```python
petal_length_values = iris_df['petallength'].astype(float)
bins = [-np.inf, 3, 5, np.inf]
labels = ['small', 'medium', 'large']
petal_length_category = pd.cut(petal_length_values, bins=bins, labels=labels)
print(petal_length_category[:10])
```

Q5 Write a Pandas function to find the most frequent value of petal length (3rd column) in iris dataset

```python
petal_length_float = iris_df['petallength'].astype(float)
most_frequent_petal_length = petal_length_float.mode()[0]
print(most_frequent_petal_length)
```
